## preprocess training data

```python
import pandas as pd

# get the training dataset
train_data = pd.read_csv("./train.csv")
```

```python
train_data.dropna(inplace=True)
train_data.reset_index(drop=True, inplace=True)
```

```python
# deal with purchase date
train_data["purchase_date"] = pd.to_datetime(train_data["purchase_date"])
# train_data['purchase_Year'] = train_data['purchase_date'].dt.year 
# train_data['purchase_Month'] = train_data['purchase_date'].dt.month 
# train_data['purchase_Day'] = train_data['purchase_date'].dt.day     
# train_data["purchase_Weekday"] = train_data['purchase_date'].dt.dayofweek

# deal with release_date
train_data["release_date"] = pd.to_datetime(train_data["release_date"])
# train_data['release_Year'] = train_data['release_date'].dt.year
# train_data['release_Month'] = train_data['release_date'].dt.month 
# train_data['release_Day'] = train_data['release_date'].dt.day       
# train_data["release_Weekday"] = train_data['release_date'].dt.dayofweek 

# get the different days between purchase_date and release_date
train_data["diff_day"] = train_data.apply(lambda raw: (raw["purchase_date"] - raw["release_date"]).days, axis=1)

# delete purchase_date and release_date
train_data.drop(columns=["purchase_date", "release_date"], inplace=True)

train_data.head()
```

```python
# add negative rate and positive rate
train_data["negative_rate"] = train_data.apply(lambda raw: raw["total_negative_reviews"]\
                                               /(raw["total_negative_reviews"] + raw["total_positive_reviews"])\
                                               if raw["total_negative_reviews"] + raw["total_positive_reviews"] != 0\
                                               else 0.5, axis=1)
train_data["positive_rate"] = train_data.apply(lambda raw: raw["total_positive_reviews"]\
                                               /(raw["total_negative_reviews"] + raw["total_positive_reviews"])\
                                               if raw["total_negative_reviews"] + raw["total_positive_reviews"] != 0\
                                               else 0.5, axis=1)
train_data["total_reviews"] = train_data.apply(lambda raw: (raw["total_negative_reviews"] + raw["total_positive_reviews"]),\
                                               axis=1)

# delete total_positive_reviews and total_negative_reviews
train_data.drop(columns=["total_positive_reviews", "total_negative_reviews"], inplace=True)
train_data.head()
```

```python
# deal with genres, categories, tags
genres = train_data["genres"].str.get_dummies(",")
categories = train_data["categories"].str.get_dummies(",") 
tags = train_data["tags"].str.get_dummies(",") 
data = pd.concat([train_data, genres, categories, tags], axis=1)
data.head()
```

```python
# delete useless columns
data.drop(columns=["id", "is_free", "genres", "categories", "tags"], inplace=True)
data.head()
```

```python
# change price and total_reviews to type(int)
data[["total_reviews", "price"]] = data[["total_reviews", "price"]].astype("int")
data.head()
```

```python
# split to data_X and data_Y
data_X = data.drop(columns=["playtime_forever"])
data_Y = data[["playtime_forever"]]
```

```python
from sklearn.model_selection import train_test_split
from math import sqrt
from sklearn.ensemble import RandomForestRegressor
import lightgbm as lgb
from sklearn.metrics import mean_squared_error
from sklearn import tree
import numpy as np
```

```python
##################### Desicion Tree model #################################
rms = 0
n = 1000
for i in range(n):
    X_train, X_test, Y_train, Y_test = train_test_split(data_X, data_Y, test_size=0.1)
    model = tree.DecisionTreeRegressor()
    model.fit(X_train,Y_train)

    test = model.predict(X_test)

    rms += sqrt(mean_squared_error(Y_test, test))
print("The rms of Decision Tree is:", rms/n)

# X_train, X_test, Y_train, Y_test = train_test_split(data_X, data_Y, test_size=0.1)
# # rf = lgb.LGBMRegressor()
# model = tree.DecisionTreeRegressor()
# model.fit(X_train,Y_train)
# test = model.predict(X_test)
# feat_labels = data_X.columns[0:]

# importances = model.feature_importances_
# indices = np.argsort(importances)[::-1]
# for f in range(X_train.shape[1]):
#     print(feat_labels[indices[f]], importances[indices[f]])
```

```python
##################### Random Forest model #################################
import warnings
warnings.filterwarnings("ignore")

rms = 0
n = 1000
for i in range(n):
    X_train, X_test, Y_train, Y_test = train_test_split(data_X, data_Y, test_size=0.1)  
    model = RandomForestRegressor(n_estimators=300, max_features="sqrt")
    model.fit(X_train,Y_train)

    test = model.predict(X_test)

    rms += sqrt(mean_squared_error(Y_test, test))
print("The rms of Random Forest is:", rms/n)

# X_train, X_test, Y_train, Y_test = train_test_split(data_X, data_Y, test_size=0.1)
# model = RandomForestRegressor()
# model.fit(X_train,Y_train)
# test = model.predict(X_test)
# feat_labels = data_X.columns[0:]

# importances = model.feature_importances_
# indices = np.argsort(importances)[::-1]
# for f in range(X_train.shape[1]):
#     print(feat_labels[indices[f]], importances[indices[f]])
```

```python
from sklearn.model_selection import GridSearchCV

X_train, X_test, Y_train, Y_test = train_test_split(data_X, data_Y, test_size=0.1) 
param_test1 = {'n_estimators':[50,75,100,125,150,175,200,225,250]}
gsearch1 = GridSearchCV(estimator = RandomForestRegressor(max_features='sqrt'), 
                       param_grid = param_test1, scoring='roc_auc',cv=5)
gsearch1.fit(X_train,Y_train)
print(gsearch1.best_params_, gsearch1.best_score_)
```

```python
##################### LightGBM model #################################
rms = 0
n = 1000
for i in range(n):
    print(i)
    X_train, X_test, Y_train, Y_test = train_test_split(data_X, data_Y, test_size=0.1)  
    model = lgb.LGBMRegressor(n_estimators=20)
    model.fit(X_train,Y_train)

    test = model.predict(X_test)

    rms += sqrt(mean_squared_error(Y_test, test))
print("The rms of LightGBM is:", rms/n)

# X_train, X_test, Y_train, Y_test = train_test_split(data_X, data_Y, test_size=0.1)
# model = lgb.LGBMRegressor()
# model.fit(X_train,Y_train)
# test = model.predict(X_test)
# feat_labels = data_X.columns[0:]

# importances = model.feature_importances_
# indices = np.argsort(importances)[::-1]
# for f in range(X_train.shape[1]):
#     print(feat_labels[indices[f]], importances[indices[f]])
```
